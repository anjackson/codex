Historical UKWA Solr Query Performance Benchmarking
---------------------------------------------------

We have indexed the Historical UK Web Archive collection (c. 30TB) into 24 Solr shards. Each is about 0.7TB of data. We have them spread over four servers, replicated to support two configurations.  

Servers 181 and 182 are identical SuperMicro models, each with 256GB of RAM, and each with 12 local hard drives mounted directly (no RAID). Each of the 12 drives contains one of the shards, and each of these servers is running 12 Solr servers, one per shard. i.e. these two machines have all the data and can be used to query the entire dataset.

However, the data was actually indexes across four servers. At that time, there were six shards on each of them (called 181, 182, 203, 215). The 203 and 215 servers are HP models, using SAN disks for storage, and have 158GB and 252GB of RAM respectively.

Each Solr service is running on a dedicated port, under Jetty, and is able to use up to 20GB of heap space (-Xmx). For 181/182, this means 12 x 20GB = 240 GB of the total 252GB available. For 203, this means 6 x 20GB = 120GB of the 158GB available. For 215, this means 6 x 20GB = 120GB of the 252GB available.

Note that, due to memory-mapped disk contention when running Solr on software RAID, we switch to using the NIOFSDirectoryFacteory, rather than the default MMAPDirectoryFactory. However, we have since stopped using software RAID, and even if we were using it, the ideal configuration for querying might be different than for indexing.

After having indexed the data on the four servers, we were planning to use just the first two for the query service, but we are seeing significant performance problems. We were not expecting particuarly stellar performance, but even then we were disappointed. It's not clear why the performance is so bad, so this is an attempt to analysis the performance.

### Measuring the performance of particular shards

Solr provides [a mechanism for specifying the shards that are to be queried](https://cwiki.apache.org/confluence/display/solr/Distributed+Requests), by passing a comma-separated lists as the ```shard``` parameter. e.g. this query will return results across two specific shards:

    http://192.168.1.215:8983/solr/jisc5/select?wt=json&indent=true&shards=192.168.1.215:8983/solr/jisc5,192.168.1.215:8984/solr/jisc5&q=*:*


Appropriate query configurations have been baked into a script, [solr-test-suite.py](https://github.com/ukwa/shine/blob/master/python/solr-test-suite.py), which picks random words/phrases from a subset of the WordNet dictionary to generate a sequence of queries.


Simple Phrase Queries
---------------------

We start by analysing all the data from simple queries. i.e. we just use "phrase search" to search for the random words and phrases. No sorting, no facets, no complicated features. We can load this data in and take a look. First, we show a simple scatter plot of the query times in milliseconds, versus six configurations:

- **181**: Run random phrase queries over a single shard, on the 181 server.
- **182**: Run random phrase queries over a single shard, on the 182 server.
- **203**: Run random phrase queries over a single shard, on the 203 server.
- **215**: Run random phrase queries over a single shard, on the 215 server.
- **2 Servers**: Run random phrase queries over all shards, but only on the 181+182 servers (12 shards on each).
- **2 Servers**: Run random phrase queries over all shards, but evenly over all servers (6 shards on each).


````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import re

def loadSB2( file ):
    ins = open( file, "r" )
    qts = []
    nhs = []
    for line in ins:
        line = re.sub("FACETS-.+\.QTime","FACETS-QTime",line)
        cols = line.split()
        if len(cols) > 1 and re.match(".*QTime.*",cols[0]) and not re.match(".*-ALL.*",cols[0]):
            qt = float(cols[1])/1000.0
            nf = int(cols[3])
            wc = float(cols[5])/1000.0
            qts.append(qt)
            nhs.append(nf)
    ins.close()
    return (np.array(qts), np.array(nhs))

qtnf181, nfnf181 = loadSB2("solr-bench/testv3-no-facets-181-only.log")
qtnf182, nfnf182 = loadSB2("solr-bench/testv3-no-facets-182-only.log")
qtnf203, nfnf203 = loadSB2("solr-bench/testv3-no-facets-203-only.log")
qtnf215, nfnf215 = loadSB2("solr-bench/testv3-no-facets-215-only.log")
qtnf2m, nfnf2m = loadSB2("solr-bench/testv3-no-facets-2servers.log")
qtnf2m2, nfnf2m2 = loadSB2("solr-bench/testv3-no-facets-2servers-2.log")
qtnf4m, nfnf4m = loadSB2("solr-bench/testv3-no-facets-4servers.log")


fig = plt.figure(figsize=(12, 6))
plt.scatter(np.repeat(1, qtnf181.shape[0]), qtnf181, label="181");
plt.scatter(np.repeat(2, qtnf182.shape[0]), qtnf182, label="182");
plt.scatter(np.repeat(3, qtnf203.shape[0]), qtnf203, label="203", color="green");
plt.scatter(np.repeat(4, qtnf215.shape[0]), qtnf215, label="215", color="lightgreen");
plt.scatter(np.repeat(5, qtnf2m.shape[0]), qtnf2m, label="2 Servers", color="red");
plt.scatter(np.repeat(5, qtnf2m2.shape[0]), qtnf2m2, color="red");
plt.scatter(np.repeat(6, qtnf4m.shape[0]), qtnf4m, label="4 Servers", color="purple");
plt.legend()
plt.ylim([0,140])
plt.ylabel("QTime [s]")

ax = plt.gca()
group_labels = ['', '181', '182', '203', '215', '2 Servers', '4 Servers']
ax.set_xticklabels(group_labels)
````

We can see that performance is very variable, with very high query times for the outliers.  203 appears to be struggling the most, and 215 appears to give the best performance. When querying over the the full dataset, the spread is significantly worse, although the average query times for the two configurations are comparable - i.e. there seems to be no advantage to running on four servers instead of two.

We now look at the query time versus the number of matching results, for individual shards:

````python
fig = plt.figure(figsize=(18, 8))
plt.scatter(nfnf181, qtnf181, label="181", color="lightblue");
plt.scatter(nfnf182, qtnf182, label="182");
plt.scatter(nfnf203, qtnf203, label="203", color="green");
plt.scatter(nfnf215, qtnf215, label="215", color="lightgreen");
plt.legend()
plt.semilogx()
plt.xlim([1e-1,1e9])
plt.ylabel("QTime [s]")
plt.xlabel("numFound")
````

There are hints of a trend here, although the performance is extremely variable. The 'peak' performance seems slightly better for 182 and 215, whereas 181 and 

Looking at the full-dataset configurations alone, we get:

````python
fig = plt.figure(figsize=(18, 8))
plt.scatter(nfnf2m, qtnf2m, label="2 Servers", color="red");
plt.scatter(nfnf2m2, qtnf2m2, color="red");
plt.scatter(nfnf4m, qtnf4m, label="4 Servers", color="purple");
plt.legend()
plt.semilogx()
plt.xlim([1e-1,10e10])
plt.ylabel("QTime [s]")
plt.xlabel("numFound")
````

Any trends appear to have been washed out. Four servers are no faster than two servers, and the outliers are pretty bad in each case.

### Summary

Given that running the same queries over four servers was no faster than over two, it seems that raw disk I/O is the main bottleneck. If the time taken to scan through one shard is fairly stable across the servers, and if the total data rate does not consume the available I/O bandwidth when scanning 12 shards in parallel on 181 or 182, spreading the same I/O load over four servers would make little difference to the performance.

However, the reason for the outliers is rather unclear. Performance is occasionally very bad, and the performance seems worse on 203, but why? is 203 oversubscribed? Is OS caching helping performance?

So, performance on 215 is significantly better than on 181 and 182 (which begin to struggle at 3/4 shards) and than 203 (which struggles at 5 shards). 

However, given that we are using the NIOFS directory factory, rathern than the memory-mapped one, and given that Solr has a fixed, large RAM allocation, it's not really clear why we should get this kind of behaviour. There should not be any RAM contention between the Solr servers themselves, so persumably it is the kernel-level filesystem caching that is under contention? Unless vanilla DocValues default to using mmap even if it is switched off at the directory-factory level?

Note that servers 181/182 have around 10GB free RAM, whereas 215 has over 100GB of free RAM. If RAM headroom is the critical issue, it is somewhat worrying that an order-of-magnitude increase in available RAM is having such a relatively modest effect.

Reading the Solr mailing lists, having too much RAM can lead to long GC pauses. Given how much RAM the queries appear to need, this seems unlikely, but cannot be ruled out just yet.
